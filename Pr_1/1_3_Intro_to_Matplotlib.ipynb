## Intro to Matplotlib
Matplotlib is the library of data visualization with Python, the official Matplotlib web page: http://matplotlib.org/

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 100)
y = x ** 2 - 3
```

### Line plot

```python
plt.figure(figsize=(5,3)) # default 6.4" × 4.8", dpi=100
plt.plot(x, y, 'r--') # red color with dashed line style
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Our Plot')
plt.grid()
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(5,3))
plt.plot(x, y1, label = 'sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.legend()
plt.show()
```

### Scatter plots

```python
np.random.seed(0)
```

```python
# generation of synthetic dataset for regression problem
num_points = 100
a = 1
b = 3
noise = np.random.normal(0, 1, num_points)
x = np.random.uniform(0, 5, num_points)
y = a * x + b + noise
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x, y, marker = '.', color = 'r')
plt.plot()
plt.plot(x, a * x + b)
plt.show()
```

### 2D scatter plot

```python
mean_1 = [1, 2]
cov_1 = [[1, 0], [0, 1]]
x_1 = np.random.multivariate_normal(mean_1, cov_1, num_points)
plt.figure(figsize=(5,3))
plt.scatter(x_1[:, 0], x_1[:, 1], marker = '^', color='coral')
plt.show()
```

```python
mean_2 = [4, 5]
cov_2 = [[2, 0], [0, 2]]
x_2 = np.random.multivariate_normal(mean_2, cov_2, num_points)
```

```python
# Gradient Fill Under the Curve

x = np.linspace(0, 10, 500)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y, color='blue')
ax.fill_between(x, y, alpha=0.1, color='blue')
ax.set_title("Gradient Fill Under the Curve")
plt.show()
```

```python
# Heatmap with Annotations
data = np.random.rand(5, 5)

fig, ax = plt.subplots(figsize=(5,3))
im = ax.imshow(data, cmap='viridis')

# Add colorbar
cbar = ax.figure.colorbar(im, ax=ax)

# Annotations
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        text = ax.text(j, i, f"{data[i, j]:.2f}",
                       ha="center", va="center", color="white")

ax.set_title("Heatmap with Annotations")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create x values
x = np.linspace(0, 2 * np.pi, 500)

# Define functions
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2 * x)
y4 = np.cos(2 * x)
y5 = np.sin(x) * np.cos(2 * x)

# Create figure and axis
fig, ax = plt.subplots(figsize=(5, 4))

colors = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']

# Plot all functions with different styles
ax.plot(x, y1, label='sin(x)', linewidth=2, color = colors[0])
ax.plot(x, y2, label='cos(x)', linewidth=2, color = colors[1])
ax.plot(x, y3, label='sin(2x)', linewidth=2, linestyle='--', color = colors[2])
ax.plot(x, y4, label='cos(2x)', linewidth=2, linestyle='-.', color = colors[3])
ax.plot(x, y5, label='sin(x)·cos(2x)', linewidth=2, linestyle=':', color = colors[4])

# Customize plot
ax.set_title('Multiple Trigonometric Functions', fontsize=16)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(loc='lower left')
ax.grid(True, alpha=0.3)

# Force display in Colab
plt.tight_layout()
plt.show()
```
